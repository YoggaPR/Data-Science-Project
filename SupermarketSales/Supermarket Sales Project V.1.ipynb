# Import and Clean Data

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r'D:\Data_Science_Project\SupermarketSales\SupermarketSalesDataset\supermarket_sales.csv')
df.head()
df.info()
df.describe()
df.isnull().sum()
```

```python
df_tes = pd.read_csv(r'D:\Data_Science_Project\SupermarketSales\SupermarketSalesDataset\supermarket_sales.csv')
count = 0
for i in range(len(df_tes['Quantity'])):
    if df_tes['Quantity'][i] <= 4:
        df_tes['Quantity'][i] = 100
    elif df_tes['Quantity'][i] <= 9:
        df_tes['Quantity'][i] = 10000
    elif df_tes['Quantity'][i] == 10:
        df_tes['Quantity'][i] = 10000000
        
print(df_tes.head(20))
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'])
df
```

# Yangon Analysis

```python
df_yangon = df[df['City'] == 'Yangon']

df_yangon = df_yangon.sort_values(by = 'Date')
df_yangon.shape

df_yangon_customer = df_yangon['Customer type'].value_counts()
# print(df_yangon_customer)

df_yangon_productbuyer = df_yangon.groupby('Product line')['Gender'].value_counts()
print(df_yangon_productbuyer)

df_yangon_meanprice = df_yangon.groupby('Product line')['Unit price'].mean()
# print(df_yangon_meanprice)

# Payment Type
df_yangon_payment = df_yangon['Payment'].value_counts()
# print(df_yangon_payment)
```

# Yangon Income Weekly

```python
df_yangon_weekly = df_yangon.resample('W', on = 'Date')
df_yangon_weekly_income = df_yangon_weekly['cogs'].sum() 
print(df_yangon_weekly_income)
```

# Yangon Income Monthly

```python
df_yangon_monthly = df_yangon.resample('M', on = 'Date')
df_yangon_monthly_income = df_yangon_monthly['cogs'].sum() 
print(df_yangon_monthly_income)
```

# Yangon Income Yearly

```python
df_yangon_yearly = df_yangon.resample('Q', on = 'Date')
df_yangon_yearly_income = df_yangon_yearly['cogs'].sum() 
print(df_yangon_yearly_income)
```

# Yangon's Cogs Daily vs Weekly vs Monthly

```python
df_yangon_daily = df_yangon.resample('D', on = 'Date')
df_yangon_daily_income = df_yangon_daily['cogs'].sum() 
df_yangon_daily_income.plot()
plt.show()

df_yangon_weekly_income.plot()
plt.show()

df_yangon_monthly_income.plot()
plt.show()
```
